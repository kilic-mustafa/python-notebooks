```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
age_list = [20,25,30,35,40,45,50,55,60]
```

```python
weight_list = [70,73,84,65,45,89,56,52,45]
```

```python
age_list_numpy = np.array(age_list)
weight_list_numpy = np.array(weight_list)
```

```python
age_list_numpy
```

```python
weight_list_numpy
```

```python
plt.plot(age_list_numpy,weight_list_numpy,"b")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.grid()
plt.title("Age - Weight Graph")
plt.show()
```

```python
my_numpy1 = np.linspace(0,15,20)
```

```python
my_numpy1
```

```python
my_numpy2 = my_numpy1 ** 3
```

```python
my_numpy2
```

```python
plt.plot(my_numpy1,my_numpy2,"g--")
```

```python
plt.subplot(1,2,1) # 1 row, 2 column, 1.graph
plt.plot(my_numpy1,my_numpy2,"r--")
plt.subplot(1,2,2)
plt.plot(my_numpy2,my_numpy1,"g*")
```

```python
my_figure2 = plt.figure()

my_axes3 = my_figure2.add_axes([0.1,0.1,0.9,0.9])
my_axes3.plot(my_numpy1,my_numpy2,"r-*")
my_axes3.set_xlabel("X Data Large")
my_axes3.set_ylabel("Y Data Large")
my_axes3.set_title("Large Graph")

my_axes2 = my_figure2.add_axes([0.25,0.5,0.3,0.3])
my_axes2.plot(my_numpy1,my_numpy2,"g^")
my_axes2.set_xlabel("X Data Small")
my_axes2.set_ylabel("Y Data Small")
my_axes2.set_title("Small Graph")
```

```python
(my_fig, my_axe) = plt.subplots(1,1)

my_numpy3 = np.linspace(0,15,20)
my_numpy4 = my_numpy3 ** 2

my_axe.plot(my_numpy3,my_numpy4, color="#5C33FF", alpha=1)
my_axe.plot(my_numpy4,my_numpy3, color="#33FF3A", alpha=0.3)
```

```python
(new_fig, new_axe) = plt.subplots()
new_axe.plot(my_numpy3, my_numpy3 + 2, color="#33ff3a", linewidth=2.4)
new_axe.plot(my_numpy3, my_numpy3 + 3, color="#f333ff", linewidth=2.4)
new_axe.plot(my_numpy3, my_numpy3 + 4, color="red", linewidth=2.4, linestyle="-.")
new_axe.plot(my_numpy3, my_numpy3 + 5, color="black", linewidth=2.4, linestyle=":")
new_axe.plot(my_numpy3, my_numpy3 + 6, color="yellow", linewidth=2.4, linestyle="--", marker="o")
new_axe.plot(my_numpy3, my_numpy3 + 7, color="orange", linewidth=2.4, linestyle="--", marker="+", markersize=15)
```
